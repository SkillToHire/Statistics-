## 1.Binomial Distribution

<pre>
Binomial distribution can be defined as the probability of a Success or Failure outcome in an experimental data or survey that is repeated multiple times.

For example, when the person tosses a coin, he assumes that he will get trail on his first toss, total possible outcomes either tail or head, if he gets tail then the result is success or else failure.
</pre>

### Example:

<pre>
70% percentage of the people who purchases bike are men, if 10 data points randomly selected who purchased the bike, find the probability that exactly there are 8 are men
From the formula, we need n and x.
Here n is total number of data points has been selected is 10
x is the number that we need to find is 7
1. Probability of selecting 7 men from 10 data points is 10c7:
           10c7 can be written has: 10! / (10-8)! * 8!
           10C7 = 45
2. p is the probability of men and q is the probability of not men
           p = 70% = 0.7, q=1-0.7=0.3.
           0.7^8= 0.0576480
3. q^(n-x)
           q = 0.3, n=10, x=8
           q^(n-x) = 0.3 ^ (10-8) =0.9
4. Multiply all steps to get binomial distribution,
    = 45*0.0576480*0.9
    = 0.2334744
Therefore, the probability of getting 8 men from 10 data points is 0.233.
</pre>

```python
#python code for binomial distribution
n=10
x=8
p=0.7
q=1-p
#### Binomial Distribution
import numpy as np
import math
ncx= math.factorial(n)/(math.factorial(n-x)*math.factorial(x))
px= np.power(p,x)
qn= np.power(q,(n-x))
prob=ncx*px*qn
print("Probability of men 8 men in 10 data points:",prob)
```

## 2.Normal Distribution

<pre>
Normal distribution is also known as Gaussian Distribution and it is a type of continuous probability distribution for a random variable, it is in bell shaped curve.

in a normal distribution of the data 34% of the data points are lying in between the mean and one standard deviation above or below the mean, since normal distribution is same for both side which in symmetrical, 68 % of the data  points are in between the one standard deviation above and one standard deviation below the mean of the data points, 95% of the data points are fall under the two standard deviation above and two standard deviation below the mean, approximately 99.6% of the data points are in between the  three standard deviation above and three standard deviation below the mean.
</pre>

```python
#python code

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
x= np.random.normal(loc=0,scale=0.8,size=100)
m=x.mean()
plt.figure(figsize=(10,6))
sns.distplot(x,hist=False,kde=False,fit=stats.gamma);
plt.plot(np.full((10,1),x.mean()),[0.00,0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40,0.45],linestyle='--',c='g')
plt.hlines(0.25,xmin=-1,xmax=+1,color='r')
plt.hlines(0.05,xmin=-2,xmax=+2,color='r')
plt.hlines(0.00,xmin=-3,xmax=+3,color='r')
plt.text(round(m,2),0.45, 'mean ='+str(round(m,2)))
plt.text(-0.75,0.26, '34%')
plt.text(0.5,0.26, '34%')
plt.text(-0.5,0.23,"  one std total 68%")
plt.text(-1.5,0.06,"47.5%")
plt.text(1.2,0.06,"47.5%")
plt.text(-0.5,0.06,"two std total 95%")
plt.text(-2,0.01, "49.8%")
plt.text(1.7,0.01, "49.8%")
plt.text(-0.5,0.01,"three std total 99.6%")
plt.title("variations of standard deviations")
plt.xlim(-3,3)
```

## 3.Percentiles,Quartiles

<b>Percentile:</b>
<pre>
A percentile is a measure used in statistics indicating the value below which a given percentage of observations in a group of observations falls.

For example, the 25th percentile is the value (or score) below which 25% of the observations may be found.
The 50th percentile is pretty much the median.
</pre>

<b>Formula:</b>
<pre>
n = (P/100) * N
P = percent of the data
N = Number of ordered data points
n = Ordinal rank (it returns the number where the P value is in the dataset)
</pre>

```python
#python code
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
%matplotlib inline

k=np.random.normal(loc=0,scale=0.8,size=100)
z=pd.DataFrame(k).describe()

plt.figure(figsize=(10,6))
sns.kdeplot(k,shade=True)

plt.plot(np.full((9,1),z.loc["25%",]),[0.00,0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40],linestyle='--',c='g',label="25% percentile")
plt.plot(np.full((9,1),z.loc["50%",]),[0.00,0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40],linestyle='--',c='b',label="50% percentile")
plt.plot(np.full((9,1),z.loc["75%",]),[0.00,0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40],linestyle='--',c='black',label="75% percentile")

plt.text(z.loc["25%",]-0.2,0.1,'25%')
plt.text(z.loc["50%",]-0.2,0.1,'50%')
plt.text(z.loc["75%",]-0.2,0.1,'75%')

plt.legend(loc=0)
plt.title("percentile calculation")
plt.show()
```

<b>Quartiles:</b>
<pre>A quartile is a type of quantile which divides the number of data points into quartiles. The first quartile, Q1, is the same as the 25th percentile, and the third quartile, Q3, is the same as the 75th percentile. The median is called both the second quartile, Q2, and the 50th percentile.</pre>

<b>Formula:</b>
<pre>To calculate the quartile, first step we need to sort the data, and then calculate different quartiles of the data.
1st Quartile = 1/4 (n+1) th term
3rd Quartile = 3/4 (n+1) th term
2nd Quartile = 3rd Quartile – 1st Quartile (Median)
</pre>

```python
#example
#python code
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(10,4))
k=np.random.normal(loc=0,scale=0.8,size=100)
z=pd.DataFrame(k).quantile([0.25,0.50,0.75])

plt.text(z.loc[0.25,],-0.1,'Q1')
plt.text(z.loc[0.50,],-0.1,'Q2')
plt.text(z.loc[0.75,],-0.1,'Q3')

sns.boxplot(k,color="green")
plt.title("Quartiles")
plt.show()
```

## 4.Inter Quartile Range

<pre>The IQR is a number that indicates how spread the middle half (the middle 50%) of the dataset is and can help determine outliers. It is the difference between the Q3 and Q1.</pre>

<b>Formula:</b>
<pre>
IQR= Q3-Q1
</pre>

<b>Note:</b>
<pre>
outliers: are those data points that fall outside from the Q1 – 1.5 x IQR and Q3 + 1.5 x IQR range.
Lower limit = Q1 – 1.5 x IQR
Upper limit = Q3 + 1.5 x IQR
</pre>
