# MEAN


The sample mean is an average and in simple words sum of all the data points in the respective variable and divided by the total no. of events (data points).

```python
#to upload image
from IPython.display import Image
Image(filename="mean formula.png",width=100,height=100)
```

### Example:

<pre>
For example, we need to know average age of the total no. of students in the class, the class has 12 students, with the respective ages:

X=8,6,8,8,7,5,7,8,8,5,5,9
We compute the sample mean by adding and dividing by the number of samples,12.
Mean = (8+6+8+8+7+5+7+8+8+5+5+9)/12
Mean = 7.0
</pre>

```python
import numpy as np
X=[8,6,8,8,7,5,7,8,8,5,5,9]

#converting a list to an numpy array
X=np.array(X)

#calculate Mean
M=X.mean()
print("Mean of X",M)
```

# MEDIAN

Median is middle score, first we need to sort the data points from smallest to largest

```python
Image(filename="median formula.png",width=500,height=500)
```

### Example:

<pre>
X = [8,6,8,8,7,5,7,8,8,5,5,9]
Before you compute the median, you need to sort the data.
A sample sort order: [5, 5, 5, 6, 7, 7, 8, 8, 8, 8, 8, 9]
Length (number of observation) of the variable = 12
    First value = 7
    Second values =8
Therefore, median is equal to average of two number:
    Median = 7+8
    Median = 7.5
</pre>

```python
import numpy as np
X=[8,6,8,8,7,5,7,8,8,5,5,9]

#converting a list to an numpy array
X=np.array(X)

#calculate Median
M=np.median(X)
print("Median of X",M)
```

# MODE

The mode of a set of data values is the value that appears most frequent.

### Example:

<pre>
A = [8,6,8,8,7,5,7,8,8,5,5,9]
From the above data variable (A), the mode is 8, because the number appears 5 times in the variable
</pre>

```python
from scipy import stats
X=[8,6,8,8,7,5,7,8,8,5,5,9]

#converting a list to an numpy array
X=np.array(X)

#calculate Mode using stats module
stats.mode(X)
```

# VARIANCE

Variance is the sum squared difference from the mean of each data point, divided by the total number of data points in the respective variable minus one.

```python
Image(filename="variance formula.png",width=300,height=300)
```

### Example:

<pre>
Use of variance, it measures how far random variables are spread out from its respective mean.
In simple words we need to see distribution of the data from its mean.
Steps Required to Calculate Variance:
1.Calculate mean of x.
2.Subtract mean from each and every data point in the respective variable
3.Square them
4.Sum of the all values
5.Take average(mean) of it
</pre>

<pre>
<b>Mathematical Explanation:</b>
    
X=[30,57,47,58,56,50,52,49,54,98,51]
Mean(X) = 54.72
</pre>

```python
Image(filename="math explanation.png",width=700,height=700)
```

<pre>
Var =2638.18/11
Var=239.83(approximately)
Hence the variance, is 239.83.
</pre>

```python
#python code
import numpy as np
X=[30,57,47,58,56,50,52,49,54,98,51]

#converting a list to an numpy array
X=np.array(X)

#calculate variance
var=X.var()
print("variance of X:",round(var,2))
```

# STANDARD DEVIATION

<pre>
It is simply, square root of variance, standard deviation is a measure of the amount of variation or dispersion of a set of values.
Difference between the variance and standard deviation, variance is square of standard deviation, for example we have dataset contains student weights values between 40 to 50, mean is 51, variance is 26.1, while standard deviation 5.10.
</pre>

<b>Note:</b>
<pre>
When you have low standard deviation, your data points tend to close to the mean, high standard deviation means that 
your data points are spread out over a wide range.
</pre>
