# CENTRAL LIMIT THEORM

It can be defined as, when we have data set with large number of random variables, if the sample random variables mean is normally distributed.

### Example:

```python
Image(filename="central tendency.png",width=300,height=300)
```

<pre>
A sample mean is the average of a random subset of a larger group of random numbers. So, if you randomly picked 10 sample data points out of 100 data points and recorded their values, the average of those 10 values would be the sample mean. You could do this many time and, since it is a random selection, the sample mean would be different each time.

What constitutes a large sample is of course subjective, but the frequently required number here is a sample size should be greater than 30. In practice, your sample might need to be a fair amount bigger, depending on a few factors.
Putting it all together the CLT just says that when you have roughly 30 or more data points in your sample, the average of those numbers is part of a bell-shaped curve. So, if you took a bunch of 30 or more sized samples and plotted them, they’d look like the normal distribution, most of the averages would fall toward the centre but you’d get a few observations towards the extremes.

CLT doesn’t make any assumptions about the distribution of your underlying data. The distribution of random samples does not need to be normally distributed in order to know that the sample means of the random subsets are normally  distributed.
</pre>

<b>code using python:</b>
<pre>
I will randomly generate 100 numbers between 0 and 90. I am trying to simulate data for 100 random samples. This data would be completely random and would not follow normal distribution. As you can see in below image:
</pre>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
%matplotlib inline

r=[random.randrange(0,90,1) for i in range(100)]
d=pd.DataFrame(r)

plt.hist(d)
plt.show()
```

<pre>
Then I would take random sample (between 30–90) of the data and take the mean of each sample.
I would then plot the histogram of mean of all the samples and it should be normally distributed.
And as you can see it does have bell curve distribution and follow normal distribution.
</pre>

```python
m=[]
for i in range(1000):
    r_sample=random.randrange(30,90)
    m.append(np.mean(random.sample(r,r_sample)))
d1=pd.DataFrame(m)
d1.hist(bins=100)
plt.title("random samples mean distribution")
plt.show()
```
