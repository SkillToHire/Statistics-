## Random Variables:

<pre>
A Random Variable is a set of possible values from a random experiment. Variable denoted with X, whose possible values are numerical outcomes of a random phenomenon.

Two types of Random Variables:
1.	Discrete Random Variables
2.	Continuous Random Variables
</pre>

### 1. Discrete Random Variables:

<pre>
Discrete random variables can take on either a finite or at most a countably infinite set of discrete values (for example integers, 0,1,2,3,4,5, ……...). 
Their probability distribution is given by a probability mass function which directly maps each value of the random variable to a probability. 
If a random variable can take only a finite number of distinct values, then it must be discrete. 
</pre>

#### Examples of discrete random variables:

<pre>
The number of male students in a class (Example: 20 which is a discrete number), 
The number of websites hits per day (Example: 100 which is a discrete number,
The number of defective light bulbs in a box of 100 (Example 10 bulbs are defective).

For example, the value of x1 takes on the probability p1, the value of x2 takes on the probability p2, and so on. The probabilities pi must satisfy two requirements: every probability pi is a number between 0 and 1, and the sum of all the probabilities is 1. (p1+p2+⋯+pk=1)

Suppose a random variable X may take k different values, with the probability that X = xi defined to be P (X = xi) = pi. The probabilities pi must satisfy the following:
1: 0 < pi < 1 for each i
2: p1 + p2 + ... + pk = 1.
</pre>

### Discrete Probability Distribution: 

<pre>
This shows the probability mass function of a discrete probability Distribution. the probabilities of the X Variables X= (1,2,3,4,5) are respectively p= (0.2,0.5,0.3,0.6,0.8). A set not containing any of these points has probability zero.
</pre>

```python
##python code
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
x=(1,2,3,4,5)
p= (0.2, 0.5, 0.3,0.6,0.8)
plt.figure(figsize=(8,4))
plt.grid(True)
plt.bar(x,p,align='center')
plt.title("Disceret Random Variables")
plt.ylabel("Probabilities")
plt.xlabel("X-Variables")
plt.show()
```

### 2. Continuous Random Variables:

<pre>
A continuous random variable is one which takes an infinite number of possible values. Continuous random variables are usually measurements. Examples include height, weight, temperature, humidity, mass, etc…….
Which have a cumulative distribution function (CDF) that is absolutely continuous.

A continuous random variable is not defined at specific values. Instead, it is defined over an interval of values (Ex: values between the interval 0 and 2) and is represented by the area under a curve (in advanced mathematics, this is known as an integral). 
The probability of observing any single value is equal to 0, since the number of values which may be assumed by the random variable is infinite. 

Suppose a random variable X may take all values over an interval of real numbers. Then the probability that X is in the set of outcomes A, P(A), is defined to be the area above A and under a curve. The curve, which represents a function p(x), must satisfy the following:
     1: The curve has no negative values (p(x) > 0 for all x)
     2: The total area under the curve is equal to 1.
A curve meeting these requirements is known as a density curve.
</pre>


### Probability Density Function: 

<pre>
This image shows the probability density function (pdf) of the normal distribution, also called Gaussian or “bell curve”, the most important continuous random distribution. As notated on the figure, the probabilities of intervals of values corresponds to the area under the curve
</pre>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
x= np.random.normal(loc=0,scale=0.8,size=100)
m=x.mean()
plt.figure(figsize=(10,6))
sns.kdeplot(x,shade=True);
plt.plot(np.full((10,1),x.mean()),[0.00,0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40,0.45],linestyle='--',c='g')
plt.hlines(0.25,xmin=-1,xmax=+1,color='r')
plt.hlines(0.05,xmin=-2,xmax=+2,color='r')
plt.hlines(0.00,xmin=-3,xmax=+3,color='r')
plt.text(round(m,2),0.45, 'mean ='+str(round(m,2)))
plt.text(-0.75,0.26, '34%')
plt.text(0.5,0.26, '34%')
plt.text(-0.5,0.23,"  one std total 68%")
plt.text(-1.5,0.06,"47.5%")
plt.text(1.2,0.06,"47.5%")
plt.text(-0.5,0.06,"two std total 95%")
plt.text(-2,0.01, "49.8%")
plt.text(1.7,0.01, "49.8%")
plt.text(-0.5,0.01,"three std total 99.6%")
plt.title("Continuous Random Variables")
plt.xlim(-3,3)
```
