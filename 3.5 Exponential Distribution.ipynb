## Exponential Distribution:

<pre>
Exponential Distribution is one of the widely used continuous distribution.
It is often used to find the time elapsed between events. In simple words the probability of an event occurring between two timeframes.
</pre>

<pre>
For example, we want to predict the following:
    The amount of time until the customer finishes browsing and purchases something in your store (success).
    How many hours an engine can go without failure;
    The processing time between electrical signals;
    The years between major floods, etc.
    The amount of time you need to wait until the bus arrives (arrival).
Note:  * e^(t) the PDF of the time until the next event happens.
X ~ Exp() Is the exponential parameter  the same as  in Poisson?

One thing that would save you from the confusion later about X ~ Exp (0.25) is to remember that that 0.25 is not a time duration, but it is an event rate, which is the same as the parameter  in a Poisson process.

For example: your blog has 500 visitors a day. That is a rate. The number of customers arriving at the store in an hour, the number of earthquakes per year, the number of car accidents in a week, the number of typos on a page, the number of hairs found in Chipotle, etc., are all rates () of the unit of time, which is the parameter of the Poisson distribution.

However, when we model the elapsed time between events, we tend to speak in terms of time instead of rate, e.g., the number of years a computer can power on without failure is 10 years (instead of saying 0.1 failure/year, which is a rate), a customer arrives every 10 minutes, major hurricanes come every 7 years, etc. When you see the terminology  mean of the exponential distribution  1/ is what it means.

Therefore, from the above example:  Exp (0.25)
It means the Poisson rate will be 0.25. During a unit time (either its a minute, hour or year), the event occurs 0.25 times on average. Converting this into time terms, it takes 4 hours (a reciprocal of 0.25) until the event occurs, assuming your unit time is an hour.
</pre>

<pre>
Given the website receives on average 3 hits per minute and the average wait time between hits is .86 minutes, it seems unlikely  but not impossible  a wait-time of 4.5 minutes for another hit would happen. From the calculation. it seems this would only occur 8.78% of the time.
</pre>

```python
#python code
#PDF of the exponential distribution
import numpy as np
def PDFExponential(lamb,x): #lamb = lambda(1/lambda(which is in possion disritbution))
    if x<=0:
        pdf=0
    else:
        pdf=lamb*np.exp(-lamb*x)
    return pdf
#CDF of the exponential distribution
def CDFExponential(lamb,x): #lamb = lambda(1/lambda(which is in possion disritbution))
    if x<=0:
        cdf=0
    else:
        cdf=1-np.exp(-lamb*x)
    return cdf
#mean of the exponential distribution
def MeanExponential(lamb):
    return 1/lamb;
def VarianceExponential(lamb):
    return (1/lamb)**2;

#CDF,PDF of the exponential distribution
#when lambda=1/5 and x=3
lamb,x=1/3,4
print("CDF of Exponential:",CDFExponential(lamb,x))
print("PDF of Exponential:",PDFExponential(lamb,x))
print("Mean of Exponential:",MeanExponential(lamb))
print("Variance of Exponential:",VarianceExponential(lamb))
```

#### Exponential is Memoryless:

<pre>
An important property of the exponential distribution is its memoryless status, meaning at any given point on the distribution it is independent of previous states. Therefore, if you need to evaluate a point given a set amount of time has passed, the initial time can be reduced to zero.
Therefore, the event has no memory of anything that happened before its current evaluation.
</pre>
