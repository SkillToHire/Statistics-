### Z-Test:

<pre>
If the population samples is normally distributed with known variance sigma square, then the test statistics for a hypothesis test concerning a single population mean.

A z-score is calculated with population parameters such as:
    population mean
    population standard deviation
    It is used to validate a hypothesis that the sample drawn belongs to the same population.
</pre>

```python
from IPython.display import Image
Image(filename="z test.png",width=400,height=400)
```

<pre>
Null Hypothesis(H0): Sample mean is same as the population mean
Alternate Hypothesis(H1):  Sample mean is not same as the population mean
</pre>
<b>Note:</b> 
<pre>
If the test statistic is lower than the critical value, accept the hypothesis or else reject the hypothesis
</pre>

#### Example:

<pre>
I have dataset of wine quality data which contains different attributes in the dataset.Here I am using attribute name density in the wine quality dataset to find the probability of our hypothesis testing using z-score.
</pre>

```python
from IPython.display import Image
Image(filename="figure.png",width=400,height=400)
```

```python
#hypothesis testing

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("winequality-red.csv")
```

```python
df.columns
```

```python
df.head()
```

```python
actual_mean=df['density'].mean()
actual_std=df['density'].std()
random_samples=df['density'].sample(n=100)
population_mean=random_samples.mean()
population_std=random_samples.std()
standard_error=population_std/np.sqrt(100)
z_score=(actual_mean-population_mean)/standard_error
print("z-score :",z_score)
```

```python
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests
ztest ,pval = stests.ztest(df['density'], x2=None, value=population_mean)
print("p-value :",float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

### T-test

<pre>
A t-test is used to compare the averages(means) of two given samples. Like a z-test, a t-test also assumes a normal distribution of the sample. A t-test is used when the population parameters (mean and standard deviation) are not known. 

There are three different types of t-tests:
Independent samples t-test which compares mean for two groups
Paired sample t-test which compares means from the same group at different times
One sample t-test which tests the mean of a single group against a known mean.
</pre>

<pre>
