<pre>
In statistical hypothesis testing, the p-value or probability value or asymptotic significance is the probability for a given statistical model that, when the null hypothesis is true, the statistical summary (such as the sample mean difference between two compared groups) would be greater than or equal to the actual observed results.
Here’s an easier definition:
If the null hypothesis is true, the P value is the probability of obtaining your sample data. 
So, if the P-value is the probability of obtaining your sample data if H0 is true, it makes sense that past a certain point, we must concede that our null hypothesis is false. 
This point is called the significance level, denoted by alpha. The significance level can be thought of as how willing you are to committing errors. For example, alpha = 0.05 means you’re willing to have a 5% error rate. 5% is a very commonly used error rate.
If the P-Value ever goes below this threshold, we reject our null hypothesis.

#reject the null hypothesis

p <= alpha # cannot reject the null hypothesis
p > alpha

The significance level is set before the statistic is calculated, so that you can’t conveniently change it after you’ve seen the results.
</pre>
