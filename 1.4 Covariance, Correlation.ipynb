## Covariance

<pre>
It is a statistical measure that shows whether two variables are related by measuring how the variables change in relation to each other. 
This is clear when you break down the word. Co- as a prefix often indicates some sort of joint action (like co-workers, co-owners, coordinate) and variance refers to variation or change.
So, covariance measures how two things change together. It tells you if there is a relationship between two things and which direction that relationship is in.
</pre>

#### Types of Covariance

<pre>
So, we know that covariance is the measure of two variables vary (or change) in a predictable way together. This could be positive covariance, meaning as one increases the other also increases, or negative covariance, meaning that as one increases the other decreases.
</pre>

<pre>
Walking through this formula, we see that the covariance of the two variables (A, B) is equal to the sum of the products of the differences of each item and the mean of its variables all divided by one less than the total number of items in the set. The A and B with an over score (line on top) represent the means of each variable.

First, you need to find the mean of each variable. Typically, we call the first mentioned variable A, so that would be A-score, and the second variable B, that would be B-score

Mean (A) = (25+35+21+67+98+67+24) / 7 = 48.14
Mean (B) = (52+10+5+98+52+36+69) / 7 = 46.0

Now, you subtract each value from its respective mean and then multiply these new values together.

</pre>

```python
Image(filename="cov table.png",width=500,height=500)
```

<pre>
The sum of X*Y, which yields the values is 1982.99
The final step divides it by (n-1) = 7-1=6
Cov (A, B) = 1983.00/6
Therefore Cov (A, B) = 330.5 (approximately)
</pre>

```python
#python code
import numpy as np
A=[25,35,21,67,98,67,24]
B =[52,10,5,98,52,36,69]
X=A-np.mean(A)
Y=B-np.mean(B)
cov=np.sum(X*Y)/(len(A)-1)
print("Covariance of A,B is : ",cov)
np.cov(A,B)
```

<pre>
The covariance of this set of data is 330.5. The number is positive, so we can state that the two variables do have a positive relationship; as the score of A rises, the score of B’s also rises.

What this doesn't tell us is how strong this relationship is.

To find the strength, we need to continue to correlation.
</pre>

## Correlation

<pre>
Covariance is interesting because it is a quantitative measurement of the relationship between two variables. Correlation between two random variables, ρ (A, B) is the covariance of the two variables normalized by the variance of each variable. This normalization cancels the units out and normalizes the measure so that it is always in the range [0, 1]:
                                                  
                               Ρ (A, B) = Cov (A, B)/ sq. Root (Var(A) *Var(B))

Can you see why correlations are a stronger measure than covariance? A correlation coefficient uses the covariance of a set and takes it one step further. It is a good thing we've already calculated the covariance of our set.
To find the divisor for this equation, we first must find the variance of each variable. Here is the formula:
</pre>

<pre>
Variance of A, Var(A) = 746.40
Variance of B, Var(B) = 911.71
Cov (A, B) =330.49
Ρ (A, B) = Cov (A, B)/sq. Root (Var(A)*Var(B))
Now, we need to calculate sq. Root (Var(A)*Var(B)): 
                       Var(A)*Var(B) = 746.40*911.71
                                             = 680510.9854227405
                       sq. Root (Var(A)*Var(B)) = 824.93
Correlation:
                                   Ρ (A, B) = 330.49/824.93
                                   Ρ (A, B) = 0.40(approximately)
Therefore, the Correlation between ρ (A, B) is 0.40(Positive Correlation). 

</pre>

```python
#python code
import numpy as np
A=[25,35,21,67,98,67,24]
B =[52,10,5,98,52,36,69]
X=A-np.mean(A)
Y=B-np.mean(B)
cov=np.sum(X*Y)/(len(A)-1)
print("Covariance of A,B is : ",cov)
#### calculate sq. Root (Var(A)*Var(B)):
varA= np.var(A)
varB= np.var(B)
sqrootofAandB=np.sqrt(varA*varB)
corr=cov/sqrootofAandB
print("Correlation of A and B is :",corr)
```
