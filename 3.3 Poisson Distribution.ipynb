#### Example for Poisson Probability:

<pre>
Here x= [1,4,6,7,10], let a supermarket will get a 15-average number of orders in the time interval afternoon 1 PM to 2 PM.
The average event rate is 15 orders in the time interval 1pm to 2pm.
P(k orders in a time interval) = (15^k) *(e^-15)/ k!
P(1 order in a time interval) = (15^1) *(e^-15)/ 1!
P(4 orders in a time interval) = (15^4) *(e^-15)/ 4!
P(6 orders in a time interval) = (15^6) *(e^-15)/ 6!
P(7 orders in a time interval) = (15^7) *(e^-15)/ 7!
P(10 orders in a time interval) = (15^10) *(e^-15)/ 10!
</pre>

```python
#python code

import numpy as np
import math
k=[1,2,3,4,5,6,8,9,10,7,11,12]
k3=[]
lambd=10
for i in range(len(k)):
    ## formula (lambda^k) *(e^-lambda)/ k!
    p=(np.power(lambd,k[i])*np.exp(-lambd))/math.factorial(k[i])
    k3.append(p)
    print("probabilty of k="+str(k[i])+" is",round(p,4))
```
