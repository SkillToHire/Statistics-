## Binomial Distribution

<pre>
Binomial distribution can be defined as the probability of a Success or Failure outcome in an experimental data or survey that is repeated multiple times.
For example, when the person tosses a coin, he assumes that he will get trail on his first toss, total possible outcomes either tail or head, if he gets tail then the result is success or else failure.
</pre>

#### example:

<pre>
70% percentage of the people who purchases bike are men, if 10 data points randomly selected who purchased the bike, find the probability that exactly there are 8 are men
    From the formula, we need n and x.
    Here n is total number of data points has been selected is 10
    x is the number that we need to find is 7
Probability of selecting 7 men from 10 data points is 10c7:
    10c7 can be written has: 10! / (10-8)! * 8!
    10C7 = 45
p is the probability of men and q is the probability of not men
     p = 70% = 0.7, q=1-0.7=0.3.
     0.7^8= 0.0576480
q^(n-x)
    q = 0.3, n=10, x=8
    q^(n-x) = 0.3 ^ (10-8) =0.9
Multiply all steps to get binomial distribution,
    = (45x0.0576480x0.9)=>0.2334744
Therefore, the probability of getting 8 men from 10 data points is 0.233.
</pre>

```python
#python code for binomial distribution
n=10
x=8
p=0.7
q=1-p
#### Binomial Distribution
import numpy as np
import math
ncx= math.factorial(n)/(math.factorial(n-x)*math.factorial(x))
px= np.power(p,x)
qn= np.power(q,(n-x))
prob=ncx*px*qn
print("Probability of men 8 men in 10 data points:",prob)
```
